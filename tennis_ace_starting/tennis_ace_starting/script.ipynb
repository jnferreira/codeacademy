```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load and investigate the data here:
df_stats = pd.read_csv('tennis_stats.csv')
df_stats.info()
```

```python
rows = 4
columns = 2

plt.figure(figsize=(15, 10))

plt.subplots_adjust(bottom =1, top = 3)

plt.subplot(rows, columns, 1)
plt.scatter(df_stats['Aces'], df_stats['Wins'])
plt.xlabel('Aces')
plt.ylabel('Wins')

plt.subplot(rows, columns, 2)
plt.scatter(df_stats['ServiceGamesPlayed'], df_stats['Wins'])
plt.xlabel('ServiceGamesPlayed')
plt.ylabel('Wins')

plt.subplot(rows, columns, 3)
plt.scatter(df_stats['BreakPointsOpportunities'], df_stats['Wins'])
plt.xlabel('BreakPointsOpportunities')
plt.ylabel('Wins')

plt.subplot(rows, columns, 4)
plt.scatter(df_stats['DoubleFaults'], df_stats['Wins'])
plt.xlabel('DoubleFaults')
plt.ylabel('Wins')

plt.subplot(rows, columns, 5)
plt.scatter(df_stats['ReturnGamesPlayed'], df_stats['Wins'])
plt.xlabel('ReturnGamesPlayed')
plt.ylabel('Wins')

plt.subplot(rows, columns, 6)
plt.scatter(df_stats['Winnings'], df_stats['Wins'])
plt.xlabel('Winnings')
plt.ylabel('Wins')

plt.subplot(rows, columns, 7)
plt.scatter(df_stats['BreakPointsOpportunities'], df_stats['Winnings'])
plt.xlabel('BreakPointsOpportunities')
plt.ylabel('Winnings')

plt.subplot(rows, columns, 8)
plt.scatter(df_stats['ServiceGamesPlayed'], df_stats['Winnings'])
plt.xlabel('ServiceGamesPlayed')
plt.ylabel('Winnings')

plt.show()
```

```python
x = np.array(df_stats['ServiceGamesPlayed']).reshape(-1, 1)
y = df_stats['Wins']
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

```python
simple_lrmodel = LinearRegression()
simple_lrmodel.fit(x_train, y_train)
simple_lrmodel.score(x_test, y_test)
```

```python
prediction = simple_lrmodel.predict(x_test)
plt.scatter(y_test, prediction, alpha=0.4)
```
